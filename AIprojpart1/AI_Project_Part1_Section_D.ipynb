```python
#GROUP MEMBERS
20L-0999
20L-2097
20L-2192



import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import random
from IPython.display import display #bcz train.head() not working in some places  so for this -> display(train.head())

# Read the file into a DataFrame: df test and train
train = pd.read_csv('/content/train - Sheet1.csv')
test = pd.read_csv('/content/test - Sheet1.csv')
#train.head()


train_copy = train.copy() #copy of train dataset
test_copy = test.copy() #copy of test dataset

train.head()

# drop multiple columns

train.drop(['address', 'famsize','Pstatus','reason','guardian','traveltime','schoolsup','famsup','nursery',
           'higher','internet','romantic','famrel','freetime','goout','Dalc','Walc'], axis=1,inplace=True)

train.drop(['Medu','Fedu','Mjob','Fjob'],axis=1,inplace=True) #axis=1 means column inplace=true 

test.drop(['address', 'famsize','Pstatus','reason','guardian','traveltime','schoolsup','famsup','nursery',
           'higher','internet','romantic','famrel','freetime','goout','Dalc','Walc'], axis=1,inplace=True)

test.drop(['Medu','Fedu','Mjob','Fjob'],axis=1,inplace=True) #axis=1 means column inplace=true 

#  school - student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
#  sex - student's sex (binary: 'F' - female or 'M' - male)
#  age - student's age (numeric: from 15 to 22)
#  guardian - student's guardian (nominal: 'mother', 'father' or 'other')
#  studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
#  failures - number of past class failures (numeric: n if 1<=n<3, else 4)
#  paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
#  activities - extra-curricular activities (binary: yes or no)
#  health - current health status (numeric: from 1 - very bad to 5 - very good)
#  absences - number of school absences (numeric: from 0 to 93)

#train.columns
#test.columns
display(train.head(7))
display(test.head(7))
#train.info()

train.columns = ['school','gender','age','studytime','failures','extraClasses','activities','health','absences','sessional-1','sessional-2','Final']
test.columns = ['school','gender','age','studytime','failures','extraClasses','activities','health','absences','sessional-1','sessional-2','Final']
display(test.head(7))
display(train.head(7))




#################################### Convert Categorial Variables to Numerical variables ####################################


train['school']=train['school'].replace(['GP'],'0')
train['school']=train['school'].replace(['MS'],'1')

test['school']=test['school'].replace(['GP'],'0')
test['school']=test['school'].replace(['MS'],'1')



train['gender']=train['gender'].replace(['M'],'0')
train['gender']=train['gender'].replace(['F'],'1')

test['gender']=test['gender'].replace(['M'],'0')
test['gender']=test['gender'].replace(['F'],'1')



train['extraClasses']=train['extraClasses'].replace(['no'],'0')
train['extraClasses']=train['extraClasses'].replace(['yes'],'1')

test['extraClasses']=test['extraClasses'].replace(['no'],'0')
test['extraClasses']=test['extraClasses'].replace(['yes'],'1')



train['activities']=train['activities'].replace(['no'],'0')
train['activities']=train['activities'].replace(['yes'],'1')

test['activities']=test['activities'].replace(['no'],'0')
test['activities']=test['activities'].replace(['yes'],'1')



#dtypes to int
train['school']=train['school'].astype(int)
test['school']=test['school'].astype(int)

train['gender']=train['gender'].astype(int)
test['gender']=test['gender'].astype(int)

train['extraClasses']=train['extraClasses'].astype(int)
test['extraClasses']=test['extraClasses'].astype(int)


train['activities']=train['activities'].astype(int)
test['activities']=test['activities'].astype(int)

train.info()
test.info()



#################################### Filling Nulls ####################################

import random
train['school'] = train['school'].fillna(random.randint(0,1))
train['gender'] = train['gender'].fillna(random.randint(0,1))
train['age'] = train['age'].fillna(random.randint(15,22))
train['studytime'] = train['studytime'].fillna(random.randint(1,4))
# train['failures'] = train['failures'].fillna(random.randint(0,4))
train['extraClasses'] = train['extraClasses'].fillna(random.randint(0,1))
train['activities'] = train['activities'].fillna(random.randint(0,1))
train['health'] = train['health'].fillna(random.randint(1,5))
train['absences'] = train['absences'].fillna(random.randint(0,93))
train['sessional-1'] = train['sessional-1'].fillna(random.randint(0,20))
train['sessional-2'] = train['sessional-1'].fillna(random.randint(1,20))
train['Final'] = train['Final'].fillna(random.randint(0,203))


test['school'] = test['school'].fillna(random.randint(0,1))
test['gender'] = test['gender'].fillna(random.randint(0,1))
test['age'] = test['age'].fillna(random.randint(15,22))
test['studytime'] = test['studytime'].fillna(random.randint(1,4))
# train['failures'] = train['failures'].fillna(random.randint(0,4))
test['extraClasses'] = test['extraClasses'].fillna(random.randint(0,1))
test['activities'] = test['activities'].fillna(random.randint(0,1))
test['health'] = test['health'].fillna(random.randint(1,5))
test['absences'] = test['absences'].fillna(random.randint(0,93))
test['sessional-1'] = test['sessional-1'].fillna(random.randint(0,20))
test['sessional-2'] = test['sessional-1'].fillna(random.randint(1,20))
test['Final'] = test['Final'].fillna(random.randint(0,203))



####################################   Normalization   ####################################


# Min-Max scaling

age_norm = (train['age'] - train['age'].min()) / (train['age'].max() - train['age'].min())
absences_norm = (train['absences'] - train['absences'].min()) / (train['absences'].max() - train['absences'].min())
health_norm = (train['health'] - train['health'].min()) / (train['health'].max() - train['health'].min())
sessional1_norm = (train['sessional-1'] - train['sessional-1'].min()) / (train['sessional-1'].max() - train['sessional-1'].min())
sessional2_norm = (train['sessional-2'] - train['sessional-2'].min()) / (train['sessional-2'].max() - train['sessional-2'].min())
final_norm = (train['Final'] - train['Final'].min()) / (train['Final'].max() - train['Final'].min())



# Print the normalized variables

print(age_norm)
train['age'] = age_norm
train['absences']=absences_norm
train['health']=health_norm
train['sessional-1']=sessional1_norm
train['sessional-2']=sessional2_norm
train['Final']=final_norm

age_norm = (test['age'] - test['age'].min()) / (test['age'].max() - test['age'].min())
absences_norm = (test['absences'] - test['absences'].min()) / (test['absences'].max() - test['absences'].min())
health_norm = (test['health'] - test['health'].min()) / (test['health'].max() - test['health'].min())
sessional1_norm = (test['sessional-1'] - test['sessional-1'].min()) / (test['sessional-1'].max() - test['sessional-1'].min())
sessional2_norm = (test['sessional-2'] - test['sessional-2'].min()) / (test['sessional-2'].max() - test['sessional-2'].min())
final_norm = (test['Final'] - test['Final'].min()) / (test['Final'].max() - test['Final'].min())


# Print the normalized age variable
print(age_norm)
test['age'] = age_norm
test['absences']=absences_norm
test['health']=health_norm
test['sessional-1']=sessional1_norm
test['sessional-2']=sessional2_norm
test['Final']=final_norm

stats = test.describe()
print(stats)

#####################################  POPULATING CSV FILES WITH PREPROCESSED DATAAAA  #####################################

train.to_csv('Preprocessed_traindata.csv')
train.to_csv('Preprocessed_testdata.csv')

```
