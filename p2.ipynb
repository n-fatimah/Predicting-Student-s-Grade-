```python

# 20L-2097
# 20L-0999
# 20L-2192

import joblib
import tkinter as tk
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import random
from IPython.display import display
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Read the file into a DataFrame: df test and train
train = pd.read_csv('train - Sheet1.csv')
test = pd.read_csv('test - Sheet1.csv')
#train.head()


train_copy = train.copy() #copy of train dataset
test_copy = test.copy() #copy of test dataset

#train.head()

# drop multiple columns

train.drop(['address', 'famsize','Pstatus','reason','guardian','traveltime','schoolsup','famsup','nursery',
           'higher','internet','romantic','famrel','freetime','goout','Dalc','Walc'], axis=1,inplace=True)

train.drop(['Medu','Fedu','Mjob','Fjob'],axis=1,inplace=True) #axis=1 means column inplace=true 

test.drop(['address', 'famsize','Pstatus','reason','guardian','traveltime','schoolsup','famsup','nursery',
           'higher','internet','romantic','famrel','freetime','goout','Dalc','Walc'], axis=1,inplace=True)

test.drop(['Medu','Fedu','Mjob','Fjob'],axis=1,inplace=True) #axis=1 means column inplace=true 

#  school - student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
#  sex - student's sex (binary: 'F' - female or 'M' - male)
#  age - student's age (numeric: from 15 to 22)
#  guardian - student's guardian (nominal: 'mother', 'father' or 'other')
#  studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
#  failures - number of past class failures (numeric: n if 1<=n<3, else 4)
#  paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
#  activities - extra-curricular activities (binary: yes or no)
#  health - current health status (numeric: from 1 - very bad to 5 - very good)
#  absences - number of school absences (numeric: from 0 to 93)

train.columns = ['school','gender','age','studytime','failures','extraClasses','activities','health','absences','sessional-1','sessional-2','Final']
test.columns = ['school','gender','age','studytime','failures','extraClasses','activities','health','absences','sessional-1','sessional-2','Final']
#display(test.head(7))
#display(train.head(7))




#################################### Convert Categorial Variables to Numerical variables ####################################


train['school']=train['school'].replace(['GP'],'0')
train['school']=train['school'].replace(['MS'],'1')

test['school']=test['school'].replace(['GP'],'0')
test['school']=test['school'].replace(['MS'],'1')



train['gender']=train['gender'].replace(['M'],'0')
train['gender']=train['gender'].replace(['F'],'1')

test['gender']=test['gender'].replace(['M'],'0')
test['gender']=test['gender'].replace(['F'],'1')

train['extraClasses']=train['extraClasses'].replace(['no'],'0')
train['extraClasses']=train['extraClasses'].replace(['yes'],'1')

test['extraClasses']=test['extraClasses'].replace(['no'],'0')
test['extraClasses']=test['extraClasses'].replace(['yes'],'1')

train['activities']=train['activities'].replace(['no'],'0')
train['activities']=train['activities'].replace(['yes'],'1')

test['activities']=test['activities'].replace(['no'],'0')
test['activities']=test['activities'].replace(['yes'],'1')



#dtypes to int
train['school']=train['school'].astype(int)
test['school']=test['school'].astype(int)

train['gender']=train['gender'].astype(int)
test['gender']=test['gender'].astype(int)

train['extraClasses']=train['extraClasses'].astype(int)
test['extraClasses']=test['extraClasses'].astype(int)


train['activities']=train['activities'].astype(int)
test['activities']=test['activities'].astype(int)

#train.info()
#test.info()



#################################### Filling Nulls ####################################

import random
train['school'] = train['school'].fillna(random.randint(0,1))
train['gender'] = train['gender'].fillna(random.randint(0,1))
train['age'] = train['age'].fillna(random.randint(15,22))
train['studytime'] = train['studytime'].fillna(random.randint(1,4))
# train['failures'] = train['failures'].fillna(random.randint(0,4))
train['extraClasses'] = train['extraClasses'].fillna(random.randint(0,1))
train['activities'] = train['activities'].fillna(random.randint(0,1))
train['health'] = train['health'].fillna(random.randint(1,5))
train['absences'] = train['absences'].fillna(random.randint(0,93))
train['sessional-1'] = train['sessional-1'].fillna(random.randint(0,20))
train['sessional-2'] = train['sessional-1'].fillna(random.randint(1,20))
train['Final'] = train['Final'].fillna(random.randint(0,203))


test['school'] = test['school'].fillna(random.randint(0,1))
test['gender'] = test['gender'].fillna(random.randint(0,1))
test['age'] = test['age'].fillna(random.randint(15,22))
test['studytime'] = test['studytime'].fillna(random.randint(1,4))
# train['failures'] = train['failures'].fillna(random.randint(0,4))
test['extraClasses'] = test['extraClasses'].fillna(random.randint(0,1))
test['activities'] = test['activities'].fillna(random.randint(0,1))
test['health'] = test['health'].fillna(random.randint(1,5))
test['absences'] = test['absences'].fillna(random.randint(0,93))
test['sessional-1'] = test['sessional-1'].fillna(random.randint(0,20))
test['sessional-2'] = test['sessional-1'].fillna(random.randint(1,20))
test['Final'] = test['Final'].fillna(random.randint(0,203))



####################################   Normalization   ####################################


# Min-Max scaling

age_norm = (train['age'] - train['age'].min()) / (train['age'].max() - train['age'].min())
absences_norm = (train['absences'] - train['absences'].min()) / (train['absences'].max() - train['absences'].min())
health_norm = (train['health'] - train['health'].min()) / (train['health'].max() - train['health'].min())
sessional1_norm = (train['sessional-1'] - train['sessional-1'].min()) / (train['sessional-1'].max() - train['sessional-1'].min())
sessional2_norm = (train['sessional-2'] - train['sessional-2'].min()) / (train['sessional-2'].max() - train['sessional-2'].min())
#final_norm = (train['Final'] - train['Final'].min()) / (train['Final'].max() - train['Final'].min())



# Print the normalized variables

#print(age_norm)
train['age'] = age_norm
train['absences']=absences_norm
train['health']=health_norm
train['sessional-1']=sessional1_norm
train['sessional-2']=sessional2_norm
#train['Final']=final_norm

age_norm = (test['age'] - test['age'].min()) / (test['age'].max() - test['age'].min())
absences_norm = (test['absences'] - test['absences'].min()) / (test['absences'].max() - test['absences'].min())
health_norm = (test['health'] - test['health'].min()) / (test['health'].max() - test['health'].min())
sessional1_norm = (test['sessional-1'] - test['sessional-1'].min()) / (test['sessional-1'].max() - test['sessional-1'].min())
sessional2_norm = (test['sessional-2'] - test['sessional-2'].min()) / (test['sessional-2'].max() - test['sessional-2'].min())
#final_norm = (test['Final'] - test['Final'].min()) / (test['Final'].max() - test['Final'].min())


# Print the normalized age variable
print(age_norm)
test['age'] = age_norm
test['absences']=absences_norm
test['health']=health_norm
test['sessional-1']=sessional1_norm
test['sessional-2']=sessional2_norm
#test['Final']=final_norm


#####################################  POPULATING CSV FILES WITH PREPROCESSED DATAAAA  #####################################

#train.to_csv('Preprocessed_traindata.csv')
#train.to_csv('Preprocessed_testdata.csv')




grade_ranges = [
    (90, 100, 4),
    (80, 89, 3),
    (70, 79, 2),
    (60, 69, 1),
    (0, 59, 0)
]

# Function to map marks to grades
def map_marks_to_grade(marks):
    for min_val, max_val, grade in grade_ranges:
        if min_val <= marks/20*100 <= max_val:
            return grade



# Separating the independent and dependent variables for the train and test sets
train_X = train.drop('Final',axis=1)
train_y = train['Final'].apply(map_marks_to_grade)  
test_X = test.drop('Final',axis=1)  
test_y = test['Final'].apply(map_marks_to_grade) 




# Multiple linear regression model
reg_model = LinearRegression()
reg_model.fit(train_X, train_y)
y_pred = reg_model.predict(test_X)

# Evaluate Errors on Multiple linear regression model
mse = mean_squared_error(test_y, y_pred)
mae = mean_absolute_error(test_y, y_pred)
r2 = r2_score(test_y, y_pred)
print("Mean squared error of Multiple linear regression model :", mse)
print("Mean absolute error of Multiple linear regression model :", mae)
print("R2 error of Multiple linear regression model :", r2, "\n")

#Calculate Accuracy, Precision, Recall and F1 score to assess the performance of the classifier
#Multiple linear regression model
mr_pred_grades =np.clip(np.round(y_pred), 0, 4).astype(int)
print("Accuracy, Precision, Recall and F1 score to assess the performance of the classifier on Multiple linear regression model")
# Accuracy
accuracy = accuracy_score(test_y, mr_pred_grades)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(test_y, mr_pred_grades, average='weighted',zero_division=1)
print("Precision:", precision)

# Recall
recall = recall_score(test_y, mr_pred_grades, average='weighted',zero_division=1)
print("Recall:", recall)

# F1 Score
f1 = f1_score(test_y, mr_pred_grades, average='weighted',zero_division=1)
print("F1 Score:", f1)







#Decision tree model
model = DecisionTreeRegressor(max_depth=3)
model.fit(train_X, train_y)
predictions = model.predict(test_X)

# Evaluate Errors on Decision tree model
msed = mean_squared_error(test_y, predictions)
maed = mean_absolute_error(test_y, predictions)
r2d = r2_score(test_y, predictions)
print('\n')
print("Mean squared error of Multiple Decision tree model :", msed)
print("Mean absolute error of Multiple Decision tree model :", maed)
print("R2 error of Multiple Decision tree model :", r2d , "\n")


dt_pred_grades =np.clip(np.round(predictions), 0, 4).astype(int)
print("Accuracy, Precision, Recall and F1 score to assess the performance of the classifier on decision tree model")

# Accuracy
accuracy = accuracy_score(test_y, dt_pred_grades)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(test_y, dt_pred_grades, average='weighted',zero_division=1)
print("Precision:", precision)

# Recall
recall = recall_score(test_y, dt_pred_grades, average='weighted',zero_division=1)
print("Recall:", recall)

# F1 Score
f1 = f1_score(test_y, dt_pred_grades, average='weighted',zero_division=1)
print("F1 Score:", f1)





def predict_grades(marks):
    
    predictions = model.predict(marks)
    print("predicted marks",predictions)
        
    return predictions

    
    
def submit_data():
    # Retrieve the input values from the GUI form
    arr=[]
    arr.append(int(school_entry.get()))
    arr.append(int(gender_entry.get()))
    arr.append(int(age_entry.get()))
    arr.append(int(studytime_entry.get()))
    arr.append(int(failure_entry.get()))
    arr.append(int(extra_classes_entry.get()))
    arr.append(int(activities_entry.get()))
    arr.append(int(health_entry.get()))
    arr.append(int(absences_entry.get()))
    arr.append(int(sessional_1_entry.get()))
    arr.append(int(sessional_2_entry.get()))

               
               
    col_names = ['school','gender','age','studytime','failures','extraClasses','activities','health','absences','sessional-1','sessional-2']
    df=pd.DataFrame()
    
    for col, val in zip(col_names, arr):
        df[col] = [val]
    df.info()
    mystr=""
    ans1=predict_grades(df)
    ans=round(float(ans1[0]))
    if ans == 0:
        mystr="D"
    elif ans == 1:
        mystr="C"
    elif ans == 2:
        mystr="B"
    elif ans == 3:
        mystr="A"
    elif ans == 4 or ans > 4:
        mystr="A+"
    result_label.configure(text=f"Predicited grade of Finals: {mystr}!")
               
               
                            
root = tk.Tk()
root.geometry("800x800")
root.configure(background='#DFD3E3')  
root.title("Data Form")

custom_font = ("Helvetica", 16)
# School
school_label = tk.Label(root, text="School:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
school_label.grid(row=0, column=0,padx=10, pady=10)
school_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
school_entry.grid(row=0, column=2,padx=10, pady=10)

# Gender
gender_label = tk.Label(root, text="Gender:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
gender_label.grid(row=1, column=0,padx=10, pady=10)
gender_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
gender_entry.grid(row=1, column=2,padx=10, pady=10)

# Age
age_label = tk.Label(root, text="Age:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
age_label.grid(row=2, column=0,padx=10, pady=10)
age_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
age_entry.grid(row=2, column=2,padx=10, pady=10)

# Study Time
studytime_label = tk.Label(root, text="Study Time:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
studytime_label.grid(row=3, column=0,padx=10, pady=10)
studytime_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
studytime_entry.grid(row=3, column=2,padx=10, pady=10)

# Failure
failure_label = tk.Label(root, text="Failure:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
failure_label.grid(row=4, column=0,padx=10, pady=10)
failure_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
failure_entry.grid(row=4, column=2,padx=10, pady=10)

# Extra Classes
extra_classes_label = tk.Label(root, text="Extra Classes:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
extra_classes_label.grid(row=5, column=0,padx=10, pady=10)
extra_classes_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
extra_classes_entry.grid(row=5, column=2,padx=10, pady=10)

# Activities
activities_label = tk.Label(root, text="Activities:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
activities_label.grid(row=6, column=0,padx=10, pady=10)
activities_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
activities_entry.grid(row=6, column=2,padx=10, pady=10)

# Health
health_label = tk.Label(root, text="Health:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
health_label.grid(row=7, column=0,padx=10, pady=10)
health_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
health_entry.grid(row=7, column=2,padx=10, pady=10)

# Absences
absences_label = tk.Label(root, text="Absences:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
absences_label.grid(row=8, column=0,padx=10, pady=10)
absences_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
absences_entry.grid(row=8, column=2,padx=10, pady=10)

# Sessional 1
sessional_1_label = tk.Label(root, text="Sessional 1:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
sessional_1_label.grid(row=9, column=0,padx=10, pady=10)
sessional_1_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
sessional_1_entry.grid(row=9, column=2,padx=10, pady=10)

# Sessional 2
sessional_2_label = tk.Label(root, text="Sessional 2:",font=custom_font, fg="#E45E9D",bg="#DFD3E3")
sessional_2_label.grid(row=10, column=0,padx=10, pady=10)
sessional_2_entry = tk.Entry(root,foreground='#728FCE', background='white', borderwidth=3, relief='groove', font=('Arial', 12))
sessional_2_entry.grid(row=10, column=2,padx=10, pady=10)



result_label = tk.Label(root, text="")
result_label.grid(row=11, column=1, columnspan=2, padx=10, pady=10)

# Submit Button
submit_button = tk.Button(root, text="Submit", command=submit_data, bg="#E45E9D", fg="#DFD3E3", font=("Arial", 14), borderwidth=2)
submit_button.grid(row=12, column=1, columnspan=2, padx=10, pady=10)

root.mainloop()
```
